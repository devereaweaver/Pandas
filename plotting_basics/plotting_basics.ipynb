## Matplotlib Basics

`matplotlib` basically has all of its plotting features in the `pyplot` module.

```python
import matplotlib.pyplot as plt
```

```python
# let's partition Anscombe's quartet DataFrame
dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
dataset_1
```

```python
# create most basic plot for dataset 1 with lines
plt.plot(dataset_1['x'], dataset_1['y'])
```

```python
# create basic plot for dataset2 with points
plt.plot(dataset_1['x'], dataset_1['y'], 'o')
```

### Figure Objects and Axes Subplots

`matplotlib` allows for the creation of subplots to place multiple plots in one figure. 

We do so using the `subplot` method which takes the number of rows, number of columns, and the subplot location (which space will the plot go into).

```python
# create entire figure for subplot 
fig1 = plt.figure()

# add subplot 1 with 2 axes and in location 1,
# subplot 2 with 2 axes and in location 2, etc.
axes1 = fig1.add_subplot(2, 2, 1)
axes2 = fig1.add_subplot(2, 2, 2)
axes3 = fig1.add_subplot(2, 2, 3)
axes4 = fig1.add_subplot(2, 2, 4)

# use the plot method on each axis to create the plots 
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

# add titles for each subplot 
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

 # add axes titles 
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes3.set_xlabel('X')
axes3.set_ylabel('Y')
axes4.set_xlabel('X')
axes4.set_ylabel('Y')

# add main title 
fig1.suptitle("Anscombe Data")

# prevent layout overlap
fig1.set_tight_layout(True)

plt.show()
```

## Statistical Graphics Using `matplotlib`

The working dataset for this section contains data on the amount of tips people leave for various variables.

```python
# import tips data from seaborn
tips = sns.load_dataset("tips")
tips.head()
```

### Univariate

#### Histograms

Values in histograms are binned and plotted to show the distribution of the variable. Let's see how to use `matplotlib` to create one.

```python
fig2 = plt.figure()
axes1 = fig2.add_subplot(1, 1, 1)

# make the histogram to check out the distribution for the 
# total_bill variable
axes1.hist(data=tips, x='total_bill', bins=10)

# labels 
fig2.suptitle("Histogram of Total Bill")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")

plt.show()
```

### Bivariate 

#### Scatter Plot

Let's check it out...

```python
scatterplot = plt.figure() 
axes1 = scatterplot.add_subplot(1, 1, 1)

# make the scatter plot 
axes1.scatter(data=tips, x='total_bill', y='tip')

# labels
axes1.set_title("Total Bill vs. Tip")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")

plt.show()
```

#### Box Plot

Plotting a discrete variable against a continuous variable. 

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

# make the box plot
axes1.boxplot(
    # specify data, need a list if plotting multiple
    x=[
        tips.loc[tips["sex"] == "Female", "tip"],  # female vs tip
        tips.loc[tips["sex"] == "Male", "tip"]  # male vs tip
    ], 
    
    # add optional label 
    labels=["Female", "Male"]
)

# labels 
axes1.set_xlabel("Sex")
axes1.set_ylabel("Tip")
axes1.set_title("Tips by Gender")

plt.show()
```

## `Seaborn`

`Seaborn` is a declarative statistical graphics plotting system. It's built on top `matplotlib` so we can still have some fine-grained control of our plots. 

The system is useful because it can easily be used from any part of the analysis stack. 

```python
# set the default seaborn context optimized for the paper print 
sns.set_context("paper")
```

### Univariate 

#### Histogram 

Plotting with `seaborn` is a bit different. We don't do all the same steps as we did when plotting with `matplotlib`. Just check it out...

```python
# create separate figure objects and add individual subplots 
hist, ax = plt.subplots()

# use seaborn to draw a histogram into the axes
sns.histplot(data=tips, x="total_bill", ax=ax)

# set title w/matplotlib notation
ax.set_title("Total Bill Histogram")

plt.show()
```

#### Kernel Density Plots 

```python
dens, ax = plt.subplots()

sns.kdeplot(data=tips, x="total_bill", ax=ax)

ax.set_title("Total Bill Density")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Unit Probability")

plt.show()
```

#### Rug Plot

1-D representation of a variable's distribution and they're typically used with other plots to enhance visualization. 

```python
# histogram with rug plot
rug, ax = plt.subplots()

# create the two plots using the same ax
sns.rugplot(data=tips, x="total_bill", ax=ax)
sns.histplot(data=tips, x="total_bill", ax=ax)

ax.set_title("Rug Plot and Histogram of Total Bill")
ax.set_xlabel("Total Bill")

plt.show()
```

#### Distribution Plots 

The `.displot()` function allows for plotting multiple univariate plots together in a single plot. 

```python
# FacetGrid object makes it eazzzzy
fig = sns.displot(data=tips, x="total_bill", kde=True, rug=True)

fig.set_axis_labels(x_var="Total Bill", y_var="Count")
fig.figure.suptitle("Distribution of Total Bill")

plt.show()
```
